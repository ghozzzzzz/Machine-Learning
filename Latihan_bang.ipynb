```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("covid_19_indonesia_time_series_all.csv")
data.head()
```

```python
data.info()
```

```python
data['Date'].unique()
```

```python
last_date= '9/16/2022'
data[data['Date']==last_date]
```

```python
unique_location = data[data['Total Cases']>0]['Location'].unique()
unique_location.sort()
unique_location
```

```python
location_total_case = []
for i in unique_location:
  location_total_case.append(data[(data['Total Cases']>0)&(data['Location']==i)]['Total Cases'].sum())
```

```python
print(set(zip(unique_location, location_total_case)))
```

```python
nan_indices = []
for i in range(len(unique_location)):
  if type(unique_location[i])==float:
    nan_indices.append[i]
```

```python
unique_location = list(unique_location)
location_total_case = list(location_total_case)
```

```python
for i in nan_indices:
  unique_location.pop(i)
  location_total_case.pop(i)

unique_location
```

```python
plt.barh(unique_location, location_total_case)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(location_total_case)
plt.legend(unique_location, loc='best')
plt.show()
```
